```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

# "Punti" Data

```python
gdf_punti = gpd.read_file("./CTRN_Vegetazione_Punti.shp")
gdf_punti.head()
```

```python
gdf_punti.groupby("DESCRIZION").MUNICIPIO.count()
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
gdf_punti.plot(ax=ax,column="DESCRIZION",markersize=2, legend=True, alpha=0.2)
```
