# Calculation of the index of how green is the walk 

Each path between A and B: 
$(A=A_0, A_1, A_2, ... A_{n-1}, B=A_n)$ can be characterized by the number of trees on it $N_{tr}$, which is number of trees in 
$R$ radius around each segment

$\sum_{0}^{N-1} N_{tr}(A_i, A_{i+1})$.

The path is also characterized by $N_{tr}^{sh}$, which is number of trees on the shortest path 

$\sum_{0}^{N-1} N^{sh}_{tr}(S_i, S_{i+1})$, 

where $S_i, S_j$ are new (!) segments of the shortest path.


Then the green index for the path is defined as 

$\sum_{i=0}^{N-1} N_{tr}(A_i, A_{i+1})/\sum_{i=0, S_0=A_0}^{N-1} N^{sh}_{tr}(S_i, S_{i+1})$

We calculate the minimum distances of the set to a using scipy.spatial.distance.cdist function


```python

import numpy as np

# generate random set of points
np.random.seed(123)
a = np.random.randint(50, size=(100, 2))

def collect(coord, sigma: float =3.0):
    """ create a small collection of points in a neighborhood of some point 
    """
    neighborhood = []
    
    x=coord[0]
    y=coord[1]

    X = int(sigma)
    for i in range(-X, X + 1):
        Y = int(pow(sigma * sigma - i * i, 1/2))
        for j in range(-Y, Y + 1):
            neighborhood.append((x + i, y + j))

    return neighborhood

rad = 10
coord_list = [collect(coord, rad) for coord in a]
coord_set = np.array(list(set([val for sublist in coord_list for val in sublist])))

from scipy.spatial import distance

dists = distance.cdist(coord_set, a).min(axis=1)

print(dists)

# if the distance is below certain radius then we choose those points.

R_totree= 10 

dist_tree = dists[dists< R_totree] # choose 
```
